```python
import pandas as pd
import scipy
from meditab.dataset import load_TOP_data, load_HINT_augmented
from sklearn.metrics import roc_auc_score, average_precision_score

output_csv = []
for phase in ['1','2','3']:
    data = load_TOP_data(return_hint=True, drop_ec=False)
    aug_data = load_HINT_augmented()
    qa_output = pd.read_csv('qa_output/hint.csv')
    len_train = len(aug_data['df_train'])
    # aug_df = pd.concat([aug_data['df_train'], aug_data['df_test']])
    aug_data['df_train']['score'] = qa_output.iloc[:len_train]['score'].values
    aug_data['df_test']['score'] = qa_output.iloc[len_train:]['score'].values
    data['df_train']['score'] = 1
    data['df_test']['score'] = 1

    data['df_train'] = data['df_train'][data['df_train']['phase'].str.contains("Phase "+phase)]
    data['df_test'] = data['df_test'][data['df_test']['phase'].str.contains("Phase "+phase)]
    aug_data['df_train'] = aug_data['df_train'][aug_data['df_train']['phase'].str.contains("Phase "+phase)]
    aug_data['df_test'] = aug_data['df_test'][aug_data['df_test']['phase'].str.contains("Phase "+phase)]
    # print(phase, data['df_train'].shape, data['df_test'].shape, aug_data['df_train'].shape, aug_data['df_test'].shape) # continue

    train_df = pd.concat([data['df_train'], aug_data['df_train']], axis=0)
    val_df = pd.concat([data['df_test'], aug_data['df_test']], axis=0)

    preds = pd.read_csv('preds/hint_phase_{}_pred.csv'.format(phase))
    preds['pred'] = preds['pred'].values * val_df['score'].values

    # evaluate the model
    y_pred = preds.groupby('nct_id').agg({'pred':'mean'}).reset_index()
    y_true = val_df[['nct_id','label','phase']].drop_duplicates().sort_values('nct_id').reset_index(drop=True)

    auc = roc_auc_score(y_true=y_true['label'].values, y_score=y_pred['pred'].values)
    prauc = average_precision_score(y_true=y_true['label'].values, y_score=y_pred['pred'].values)
    # print('{} Test ROCAUC: {} PRAUC: {} '.format(phase, auc, prauc))
    output_csv.append('{} Test ROCAUC: {} PRAUC: {} '.format(phase, auc, prauc))

print(output_csv)
# ['1 Test ROCAUC: 0.8364571428571428 PRAUC: 0.818262256446653 ', '2 Test ROCAUC: 0.844078995978425 PRAUC: 0.7769618646699158 ', '3 Test ROCAUC: 0.8370043000477783 PRAUC: 0.5540942301307455 ']
```

```python
import json
import numpy as np
import pandas as pd

pmc_path='/srv/local/data/PMC-Patients/datasets/PMC-Patients.json'
preds_path = '/srv/local/data/chufan2/DistillTab/preds/round_0_pred.csv'
shap_value_path = '/srv/local/data/chufan2/DistillTab/preds/shap_values_round_0.npy'

PMC_Patients_json = json.load(open(pmc_path, 'r'))
all_preds = pd.read_csv(preds_path)
shap_value = np.load(shap_value_path)
num_train = len(all_preds) - len(PMC_Patients_json)
pmc_preds = all_preds.iloc[num_train:]

print(all_preds.shape, pmc_preds.shape, shap_value.shape, pmc_preds.columns)
pmc_preds['shap_value'] = shap_value
best_pos_pids = pmc_preds[pmc_preds['pred'].values > .5].sort_values('shap_value', ascending=False)['pid']
best_neg_pids = pmc_preds[pmc_preds['pred'].values <= .5].sort_values('shap_value', ascending=False)['pid']
print(best_pos_pids.shape, best_neg_pids.shape)
best_pids = best_pos_pids.tolist()[:num_train//2] + best_neg_pids.tolist()[:num_train//2]


pmc_preds['sentence'] = [PMC_Patients_json[i]['patient'] for i in range(len(PMC_Patients_json))]
pmc_preds['label'] = pmc_preds['pred'] > .5
pmc_data = pmc_preds[pmc_preds['pid'].isin(best_pids)]
```

```python
# # load the PMC patient data
# pmc_data = load_pmc_data()

# set hard labels for pmc_data
avg_mortality = 0.19299616422787327

def select_pmc_data(df, quantile=0.75):
    if quantile > 0:
        qt = df['shap_value'].quantile(quantile)
    else:
        qt = 0
    df_ = df[df['shap_value'] > qt].reset_index(drop=True)
    df_['label'] = df_['label'].apply(lambda x: 1.0 if x > avg_mortality else 0.0)
    return df_

pmc_df = select_pmc_data(pmc_data, quantile=0.5)
print("PMC label value counts:", pmc_df['label'].value_counts())
print((pmc_data['pred'] > avg_mortality).astype(float).mean())
```
